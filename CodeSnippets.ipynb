```python
def foo(**kargs):
    print(kargs)
    for i, j in kargs.items():
        print("{0} = {1}".format(i,j))
        
foo(a = "Hello", b = 70, abc = "why", wer = "far away")
```

```python
def foo(*args, **kargs):
    print(args)
    print(kargs)
    for i in args:
        print(i)
    for k, v in kargs.items():
        print("{0} = {1}".format(k, v))

foo(30, 40, 100, 10, "hello", key=abs, an="This", bc=889)
```

```python
def sqrt(x):
    if not isinstance(x, (int, float)):
        raise TypeError("Input has to be a number.")
    elif x < 0:
        raise ValueError("Input has to be positive.")
    return x**0.5
```

```python
sqrt(9)
```

```python
def factors(n):            #Generator that computes factors.
    k = 1
    while k * k < n:       #While k < sqrt(n)
        if n % k == 0:
            yield k
            yield n // k
        k += 1
        if k * k == n:     #Special case if n is perfect square.
            yield k
```

```python
factors(9)
```

```python
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a+b
```

```python
fibonacci()
```

```python
from turtle import *
a = Turtle()

def rectangle(x, y, width, height):
    a.penup()
    a.goto(x,y)
    a.pendown()
    a.setheading(9)
    for i in range(4):
        a.forward(width)
        a.forward(height)
        a.left(90)
```

```python
rectangle(-140, 0, 10, 20)
```
